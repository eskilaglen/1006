```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp

N, A = sp.symbols('N A', positive=True, real=True)
```

```python
#produktfn

def prod(c,N,A):
    produksjon=1000*(c.exp(N)/(A+c.exp(N)))
    return produksjon
prod(sp,N,A)
```

```python
#definerer funksjonene
def gjen(N,A):
    produksjon=1000*(sp.exp(N)/(A+sp.exp(N)))
    return produksjon/N

def grense(N,A):
    produksjon=1000*(sp.exp(N)/(A+sp.exp(N)))
    return sp.diff(produksjon, N)


#endrer for numpy
gjennomsnitt = sp.lambdify((N,A), gjen(N,A))
grenseverdi = sp.lambdify((N,A), grense(N,A))


verdi = 200
n=np.linspace(1,10,100)


#finner ekstremalpunkt 
grense1 = grenseverdi(n, verdi)
grense2 = np.argmax(grense1)
grenseverdi_x = n[grense2]
grenseverdi_y = grense1[grense2]
```

```python
fig1, ax = plt.subplots()
ax.set_ylabel('Kvantum (x)', loc='top')
ax.set_xlabel('Enheter arbeidskraft 👎', loc='right')
ax.set(xlim=(1,10))
ax.set(ylim=(0,400))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')    

ax.plot(n, gjennomsnitt(n,verdi), label='gjennomsnitt')
ax.plot(n,  grenseverdi(n,verdi), label='grense')
ax.scatter(grenseverdi_x,grenseverdi_y, color='red', marker='o', label='toppunkt')

ax.set_title('Produksjon ')
```
