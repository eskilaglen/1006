```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")
```

```python
df
```

```python
col_names = list(df.columns)
col_names

nydata = df[df.columns[0:4]]
nydata1 = nydata[['Year', 'log q', 'log p', 'log h']]


nydata2 = nydata1[nydata1.columns[0:3]]



p  = np.exp(nydata['log p'])
q  = np.exp(nydata['log q'])

nydata2['p'] = p
nydata2['q'] = q
nydata2
```

```python


plt.legend(['Produktfunksjon'])
plt.plot(nydata2['Year'], nydata2['p'])

plt.legend(['Produktfunksjon'])

```

```python
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(nydata2['Year'], nydata2['p'], color = "purple", label="price")
ax2.plot(nydata2['Year'], nydata2['q'], color = "red", label="quantity")
ax1.legend()
ax2.legend()
```

```python
Q = np.arange(20,105,5)
logQ = np.log(Q)

df = {'Q':Q, 'logQ':logQ}
df1 = pd.DataFrame(df)

supply = -2 + 1.7*df1['logQ']
demand = 8.5-0.82*df1['logQ']

df2 = {'supply(logP)':supply, 'demand(logP)':demand}

df2 = pd.DataFrame(df2)

df3  = pd.concat([df1, df2],axis=1)

df3['supply(P)'] = np.exp(df3['supply(logP)'])

df3['demand(P)'] = np.exp(df3['demand(logP)'])

df3
```

```python
plt.plot(df3['Q'], df3['demand(P)'], label = 'Etterspørsel')
plt.plot(df3['Q'], df3['supply(P)'], label = 'Tilbud')
plt.legend()
plt.show()
```

```python
logshocksupply = -2 + 1.7*df1['logQ']+0.4

df3['shocksupply'] = np.exp(logshocksupply)

plt.plot(df3['Q'], df3['demand(P)'])
plt.plot(df3['Q'], df3['supply(P)'])
plt.plot(df3['Q'], df3['shocksupply'])
```

```python
#utregning av priselastisitet supply
import sympy as sp
logP, logQ = sp.symbols('logP logQ')
supply = sp.Eq(-2 + 1.7*logQ, logP)
sp.solve(supply, logQ)
```

```python
#utregning av priselastisitet demand
demand = sp.Eq(8.5-0.82*logQ, logP)
sp.solve(demand, logQ)
```
